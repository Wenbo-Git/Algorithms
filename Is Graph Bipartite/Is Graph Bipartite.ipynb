#### https://leetcode.com/problems/is-graph-bipartite

```python
class Solution:
    def isBipartite(self, graph) -> bool:
        colors = {}
        
        def dfs(graph, i):
            for neighbor in graph[i]:
                if neighbor not in colors:
                    colors[neighbor] = "b" if colors[i] == "w" else "w"
                    if not dfs(graph, neighbor):
                        return False
                elif colors[neighbor] == colors[i]:
                    return False
            return True
        
        for i in range(len(graph)):
            if i not in colors:
                colors[i] = "w"
                if not dfs(graph, i):
                    return False        
        return True
```

```python
print(Solution().isBipartite([[1,3],[0,2],[1,3],[0,2]]))
```
